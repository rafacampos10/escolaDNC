## **Configuração dos Dados**

---



```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel("/content/drive/MyDrive/Escola DNC - Formação de Dados/Analise de Dados/dataset/varejo.xlsx")
vendas.head()
```

```python
cliente = pd.read_excel("/content/drive/MyDrive/Escola DNC - Formação de Dados/Analise de Dados/dataset/cliente_varejo.xlsx")
cliente.head()
```

```python
cliente.info()
```

```python
vendas.info()
```

```python
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP", "Aplicativo") # o replace substitui o nome das informações dentro da coluna
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ", "_") # o replace substitui o nome das informações dentro da coluna
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(",", "") # o replace substitui o nome das informações dentro da coluna
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas[vendas["estado"].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna() #deleta a linha toda dos nulos
vendas_sem_nulos.info()
```

```python
vendas["estado"] = vendas["estado"].fillna("MS") # coloca os dados nulos como MS
vendas["estado"].fillna("MS", inplace = True)
vendas.info()
```

```python
media_preco = vendas["Preço"].mean()
vendas["Preço"].fillna(media_preco, inplace=True)
vendas.info()
```

```python
vendas[["Preço", "Preço_com_frete", ]].describe()
```

```python
preco_errado = vendas.query("Preço > Preço_com_frete")
vendas_correto = vendas.query("Preço < Preço_com_frete")
```

```python
vendas_correto.describe()
```

```python
vendas_correto.head()
```

```python
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado =='SP'").describe()
```

```python
vendas_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending= False).reset_index() #métrica do Departamento mais vendido
```

```python
round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(), 2) #métrica do Departamento mais vendido por preço com frete
#round seria quantas casas decimais seria
```

```python
vendas_correto.groupby("Data").idcompra.nunique().sort_values(ascending= False).reset_index()
```

```python
vendas_correto["mes"] = vendas_correto["Data"].dt.month #criar uma nova coluna
vendas_correto.head()
```

```python
vendas_correto.groupby("mes").idcompra.nunique().sort_values(ascending= False).reset_index()
```

```python
cliente[["idade", "renda"]].describe()
```

```python
cliente = cliente.astype({"renda":"float"})
cliente.info()
```

```python
vendas_clientes = vendas_correto.merge(cliente, how="left", on="cliente_Log") #left join
vendas_clientes.head()
```

```python
agg_idcanal_renda = round(vendas_clientes.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending= False).reset_index(),2) #group pelo canal de venda com base na media da renda
agg_idcanal_renda
```

```python
agg_bandeira_idade = round(vendas_clientes.groupby("bandeira")["idade"].agg("mean").sort_values(ascending= False).reset_index(),2)
agg_bandeira_idade
```

## **Fazendo a parte Gráfica**

---





```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(5,4)) #ajustando o tamanho da figura

plt.bar(agg_bandeira_idade["bandeira"], agg_bandeira_idade["idade"]) #grafico de barra o bandeira é o X e a idade é o Y

plt.ylabel("Média de idade");
plt.title("Idade Média por Bandeira",loc="center", fontsize=18, fontweight="bold"); #coloca titulo no gráfico
```

```python
fig = plt.figure(figsize=(5,4)) #ajustando o tamanho da figura

plt.bar(agg_idcanal_renda["idcanalvenda"], agg_idcanal_renda["renda"]) #grafico de barra o idcanalvenda é o X e a renda é o Y

plt.ylabel("Média de renda");
plt.title("Renda Média por Canal de Venda",loc="center", fontsize=18, fontweight="bold"); #coloca titulo no gráfico
```

```python
venda_por_data = vendas_correto.groupby("Data").idcompra.nunique().reset_index()
venda_por_data
```

```python
fig = plt.figure(figsize=(15,4)) #ajustando o tamanho da figura

plt.xlabel("Data da compra")
plt.ylabel("Qtde de vendas")
x = venda_por_data["Data"]
y = venda_por_data["idcompra"]

plt.plot(x,y, linewidth ="2") #fazer grafico de linha
plt.show()
```

```python
agg_dept_preco = round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(), 2)
```

```python
fig = plt.figure(figsize=(30,20)) #ajustando o tamanho da figura

plt.bar(agg_dept_preco["Nome_Departamento"], agg_dept_preco["Preço_com_frete"]) #grafico de barra o idcanalvenda é o X e a renda é o Y

plt.ylabel("Média de Preço com Frete");
plt.title("Média de Preço com Frete por Departamento",loc="center", fontsize=18, fontweight="bold"); #coloca titulo no gráfico
```
